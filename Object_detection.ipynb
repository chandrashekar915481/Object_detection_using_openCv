```python
def detect_face(img):
    face_rects=face_classifier.detectMultiScale(img)
    
    for (x,y,w,h) in face_rects:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),10)
    
    return img
```

```python
def eye_detector(img):
    eyes_classifier=cv.CascadeClassifier('haarcascade_eye.xml')
    eyes=eyes_classifier.detectMultiScale(img)
    
    for (x,y,w,h) in eyes:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),10)
        
    return img    
```

# 3. Pedestrian Detection

```python
pedes_classifier=cv.CascadeClassifier('haarcascade_fullbody.xml')
```

```python
capture=cv.VideoCapture('Pexels Videos 2670.mp4')
```

```python
imshowSize = (640, 480)
in1='outputvideo.mp4'
WRITE_VIDEO_FLAG='out1.avi'
if WRITE_VIDEO_FLAG:
    # Define the codec and create VideoWriter object
    w = int(capture.get(4))
    h = int(capture.get(3))
    fourcc = cv.VideoWriter_fourcc(*'XVID')
    writer = cv.VideoWriter('out1.avi', fourcc, 25, imshowSize)
while True:
    isTrue,frame=capture.read()
    
    pedes=pedes_classifier.detectMultiScale(frame)
    
    if isTrue:
        for (x,y,w,h) in pedes:
            cv.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),10)
            cv.imshow('Pedestrians',frame)
            writer.write(cv.resize(frame,imshowSize))
        if cv.waitKey(1)==27 :
            break
    else:
        break

capture.release()
cv.destroyAllWindows()
```

#  4. Car Detector

```python
car_classifier=cv.CascadeClassifier('haarcascade_car.xml')
```

```python
capture=cv.VideoCapture('production ID_4174761.mp4')
```

```python
imshowSize = (640, 480)
in1='outputvideo.mp4'
WRITE_VIDEO_FLAG='out2.avi'
if WRITE_VIDEO_FLAG:
    # Define the codec and create VideoWriter object
    w = int(capture.get(4))
    h = int(capture.get(3))
    fourcc = cv.VideoWriter_fourcc(*'XVID')
    writer = cv.VideoWriter('out2.avi', fourcc, 25, imshowSize)
while True:
    isTrue,frame=capture.read()
    car=car_classifier.detectMultiScale(frame)
    
    if isTrue:
        for (x,y,w,h) in car:
            cv.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),8)
            cv.imshow('Video',frame)
            writer.write(cv.resize(frame,imshowSize))
        if cv.waitKey(1)==27:
            break
        
    else:
        break
        
capture.release()
cv.destroyAllWindows()
```

#  6. Car Plate Detector

```python
img=cv.imread('car_plate.jpg')
img_c=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img_c)
```

```python
carplate_cassifier=cv.CascadeClassifier('haarcascade_plate_number.xml')
```

```python
def car_plate(img):
    carplate_cassifier=cv.CascadeClassifier('haarcascade_plate_number.xml')
    car_plate=carplate_cassifier.detectMultiScale(img)
    
    for (x,y,w,h) in car_plate:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),5)
    
    return img
    
```

```python
result=car_plate(img_c)
plt.imshow(result)
```
